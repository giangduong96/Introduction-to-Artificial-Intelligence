```python
# Name: Giang Duong
# CS 156 SPRING 2021
# Professor Sanjoy Paul
# Homework 3
import numpy as np
import pandas as pd

# import data and data processing
filecsv = pd.read_csv("Data-Hw3.csv")
# print(filecsv)
frame = pd.DataFrame(filecsv)
print("Table values : \n", frame)
# ilocate the values such as sex, age, sibsp, parchm pclass
X = frame.iloc[:, 2:-1].values
# ilocate the survive or not
Y = frame.iloc[:, -1].values
# Y = Y.reshape(-1, 1)
# print(X)
# print(Y)

s1 = np.array(['male', 4, 0, 0, 3])
s2 = np.array(['male', 4, 4, 0, 3])
s3 = np.array(['male', 4, 0, 5, 3])
s4 = np.array(['male', 4, 0, 0, 1])
s5 = np.array(['male', 40, 0, 0, 3])
s6 = np.array(['male', 40, 4, 0, 3])
s7 = np.array(['male', 40, 0, 5, 3])
s8 = np.array(['male', 40, 0, 0, 1])
s9 = np.array(['female', 4, 0, 0, 3])
s10 = np.array(['female', 4, 4, 0, 3])
s11 = np.array(['female', 4, 0, 5, 3])
s12 = np.array(['female', 4, 0, 0, 1])
s13 = np.array(['female', 40, 0, 0, 3])
s14 = np.array(['female', 40, 4, 0, 3])
s15 = np.array(['female', 40, 0, 5, 3])
s16 = np.array(['female', 40, 0, 0, 1])
test_set = np.array([['male', 4, 0, 0, 3],
                     ['male', 4, 4, 0, 3],
                     ['male', 4, 0, 5, 3],
                     ['male', 4, 0, 0, 1],
                     ['male', 40, 0, 0, 3],
                     ['male', 40, 4, 0, 3],
                     ['male', 40, 0, 5, 3],
                     ['male', 40, 0, 0, 1],
                     ['female', 4, 0, 0, 3],
                     ['female', 4, 4, 0, 3],
                     ['female', 4, 0, 5, 3],
                     ['female', 4, 0, 0, 1],
                     ['female', 40, 0, 0, 3],
                     ['female', 40, 4, 0, 3],
                     ['female', 40, 0, 5, 3],
                     ['female', 40, 0, 0, 1]]
                    )


def show_confusion_matrix():
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(cm)
    ax.grid(False)
    ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
    ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
    ax.set_ylim(1.5, -0.5)
    for i in range(2):
        for j in range(2):
            ax.text(j, i, cm[i, j], ha='center', va='center', color='red')
    plt.title("Confusion matrix")
    plt.show()


# print(test_set)
# convert categories data: Sex
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(dtype='int'), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
test_set = np.array(ct.fit_transform(test_set))
test_set = np.array(test_set, dtype='int')
print(test_set)

# replace the missing data to make data set complete
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X = imputer.fit_transform(X)
X = imputer.transform(X)
# print(X)

# divide the data set into training set and test set
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=1)
# print("X_test\n", x_test)
# print("X_train\n", x_train)
# print("y_test\n", y_test)
# print("y_train\n", y_train)
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
test_set = sc.transform(test_set)

print("Part B: Use Logistic Regression to predict if a passenger in the Test set will survive the accident\n")
from sklearn.linear_model import LogisticRegression

logistic_Regression = LogisticRegression(solver='liblinear', random_state=0)
logistic_Regression.fit(x_train, y_train)
# predict
y_predict_logistic = logistic_Regression.predict(x_test)
y_print = np.append(y_predict_logistic, y_test).reshape(-1, 2)
print("the prediction and the corresponding ground truth in the Test set\n", y_print)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, r2_score

cm = confusion_matrix(y_test, y_predict_logistic)
print("The Confusion Matrix\n", cm)
accuracy_score_logistic = accuracy_score(y_test, y_predict_logistic)
print("Compute Accuracy", accuracy_score_logistic)
score_logistic = logistic_Regression.score(x_test, y_test)
print("Score = ", score_logistic)
print("Predict 16 cases test set after encoder and fit into the model.")
predict_logistic = logistic_Regression.predict(test_set)
print(predict_logistic)

final = pd.DataFrame(np.array(np.arange(1, 17, 1).reshape(-1, 1)), columns=["Case"])
final['logistic'] = predict_logistic
show_confusion_matrix()

print(
    "Part C: Use K Nearest Neighbor Classification with 7 neighbors to predict if a passenger in the Test set will survive the accident")
from sklearn.neighbors import KNeighborsClassifier

k_classifier = KNeighborsClassifier(n_neighbors=7, metric="minkowski", p=2)
# p =2 is order 2 each xi- yi then order 1/2 them all
k_classifier.fit(x_train, y_train)
# predict
y_predict_K_Nearest = k_classifier.predict(x_test)
print("the prediction and the corresponding ground truth in the Test set\n", y_print)

# making confusion matrix
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, r2_score

cm = confusion_matrix(y_test, y_predict_K_Nearest)
print("The Confusion Matrix\n", cm)
accuracy_score_K_Nearest = accuracy_score(y_test, y_predict_K_Nearest)
print("Compute Accuracy", accuracy_score_K_Nearest)
score_K_Nearest = k_classifier.score(x_test, y_test)
print("Score = ", score_K_Nearest)
print("Predict 16 cases test set after encoder and fit into the model.")
predict_K_Nearest = k_classifier.predict(test_set)
print(predict_K_Nearest)
final['K_Nearest'] = predict_K_Nearest
show_confusion_matrix()


print(
    "Part D: Use Support Vector Machine (SVM) Classification to predict if a passenger in the Test set will survive the accident\n")
from sklearn.svm import SVC

svm_classify = SVC(kernel="linear", random_state=0)
svm_classify.fit(x_train, y_train)
# predict
y_predict_SVM = svm_classify.predict(x_test)

y_print = np.append(y_predict_SVM, y_test).reshape(-1, 2)
print("the prediction and the corresponding ground truth in the Test set\n", y_print)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, r2_score

cm = confusion_matrix(y_test, y_predict_SVM)
print("The Confusion Matrix\n", cm)
accuracy_score_SVM = accuracy_score(y_test, y_predict_SVM)
print("Compute Accuracy", accuracy_score_SVM)
score_SVM = svm_classify.score(x_test, y_test)
print("Score = ", score_SVM)
print("Predict 16 cases test set after encoder and fit into the model.")
predict_SVM = svm_classify.predict(test_set)
print(predict_SVM)
final[' SVM '] = predict_SVM
show_confusion_matrix()

from sklearn.svm import SVC

print(
    "Part E: Kernel Support Vector Machine (K-SVM) Classification to predict if a passenger in the Test set will survive the accident\n")
svm_classify_RBF = SVC(kernel="rbf", random_state=0)
svm_classify_RBF.fit(x_train, y_train)
# predict
y_predict_K_SVM: object = svm_classify_RBF.predict(x_test)

y_print = np.append(y_predict_K_SVM, y_test).reshape(-1, 2)
print("the prediction and the corresponding ground truth in the Test set\n", y_print)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, r2_score

cm = confusion_matrix(y_test, y_predict_K_SVM)
print("The Confusion Matrix\n", cm)
accuracy_score_K_SVM = accuracy_score(y_test, y_predict_K_SVM)
print("Compute Accuracy", accuracy_score_K_SVM)
score_K_SVM = svm_classify_RBF.score(x_test, y_test)
print("Score = ", score_K_SVM)
print("Predict 16 cases test set after encoder and fit into the model.")
predict_K_SVM = svm_classify_RBF.predict(test_set)
print(predict_K_SVM)
final['K_SVM'] = predict_K_SVM
show_confusion_matrix()


from sklearn.naive_bayes import GaussianNB

print("Part F: Na√Øve Bayes Classification to predict if a passenger in the Test set will survive the accident\n")
naive_bayes_classify = GaussianNB()
naive_bayes_classify.fit(x_train, y_train)
# predict
y_predict_naive_bayes = naive_bayes_classify.predict(x_test)

y_print_naive_bayes = np.append(y_predict_naive_bayes, y_test).reshape(-1, 2)
print("the prediction and the corresponding ground truth in the Test set\n", y_print)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, r2_score

cm = confusion_matrix(y_test, y_predict_naive_bayes)
print("The Confusion Matrix\n", cm)
accuracy_score_naive_bayes = accuracy_score(y_test, y_predict_naive_bayes)
print("Compute Accuracy", accuracy_score_naive_bayes)
score_naive_bayes = naive_bayes_classify.score(x_test, y_test)
print("Score = ", score_naive_bayes)
print("Predict 16 cases test set after encoder and fit into the model.")
predict_naive_bayes = naive_bayes_classify.predict(test_set)
print(predict_naive_bayes)
final['naive_bayes'] = predict_naive_bayes
show_confusion_matrix()

##### Decision tree classification
from sklearn.tree import DecisionTreeClassifier

print("Part F: Decision tree classification to predict if a passenger in the Test set will survive the accident\n")
decision_tree_classify = DecisionTreeClassifier(criterion="entropy", random_state=0)
decision_tree_classify.fit(x_train, y_train)
# predict
y_predict_decision_tree = decision_tree_classify.predict(x_test)

y_print_decision_tree = np.append(y_predict_decision_tree, y_test).reshape(-1, 2)
print("the prediction and the corresponding ground truth in the Test set\n", y_print)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, r2_score

cm = confusion_matrix(y_test, y_predict_decision_tree)
print("The Confusion Matrix\n", cm)
accuracy_score_decision_tree = accuracy_score(y_test, y_predict_decision_tree)
print("Compute Accuracy", accuracy_score_decision_tree)
score_decision_tree = decision_tree_classify.score(x_test, y_test)
print("Score = ", score_decision_tree)
print("Predict 16 cases test set after encoder and fit into the model.")
predict_decision_tree = decision_tree_classify.predict(test_set)
print(predict_decision_tree)
final['decision_tree'] = predict_decision_tree
show_confusion_matrix()

# Random forest
from sklearn.ensemble import RandomForestClassifier

print("Part H: Random forest classification to predict if a passenger in the Test set will survive the accident\n")

random_forest_classify = RandomForestClassifier(n_estimators=10, criterion="entropy", random_state=0)
random_forest_classify.fit(x_train, y_train)
# predict
y_predict_random_forest = random_forest_classify.predict(x_test)

y_print_random_forest = np.append(y_predict_random_forest, y_test).reshape(-1, 2)
print("the prediction and the corresponding ground truth in the Test set\n", y_print)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, r2_score

cm = confusion_matrix(y_test, y_predict_random_forest)
print("The Confusion Matrix\n", cm)
accuracy_score_random_forest = accuracy_score(y_test, y_predict_random_forest)
print("Compute Accuracy", accuracy_score_random_forest)
score_random_forest = random_forest_classify.score(x_test, y_test)
print("Score = ", score_random_forest)
print("Predict 16 cases test set after encoder and fit into the model.")
predict_random_forest = random_forest_classify.predict(test_set)
print(predict_random_forest)
final['Random Forest'] = predict_random_forest

show_confusion_matrix()
final.loc[len(final.index)] = ["Accuracy Score", score_logistic, score_K_Nearest, score_SVM, score_K_SVM,
                               score_naive_bayes, score_decision_tree, score_random_forest]
print(final)
print("From the accuracy score we can see that Logistic Regression, K-SVM and Naive Bayes are the best predictive models.\n ")
print("The size of the data set, how many time of the data set get folded (n-fold), the "
      "% of split data set and test set. \n"
      "Each model is saved after running a machine learning algorithm on training data and \n"
      "represents the rules, numbers, and any other algorithm-specific data structures required to make predictions.")
```
